# gym - copy-v0

```python
env1 = environments.environment_gym('Copy-v0')
env = environments.env_wrapper(env1)
n_actions = 20
```

```python
#random policy
policy = random()
Q = defaultdict(lambda: np.zeros(n_actions))
mean_scores, score = play(env, Q, 10000, policy)
window = 100
moving_mean = []
for i in range(100,len(score)):
    moving_mean.append(np.mean(score[i-100:i]))
print('max',max(moving_mean))
plt.plot(moving_mean)
plt.show()
```

```python
policy = random()
Q_class = update_vanilla_Q(n_actions, gamma = 1.0)
mr = MC(env, policy, Q_class,500000, generate_episode, stopping = 25)
```

```python
policy = maximum()
mean_scores, score = play(env, Q_class.Q, 10000, policy)
window = 100
moving_mean = []
for i in range(100,len(score)):
    moving_mean.append(np.mean(score[i-100:i]))
print('max',max(moving_mean))
plt.plot(moving_mean)
plt.show()
```
