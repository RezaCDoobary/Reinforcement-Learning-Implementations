```python
import environments
import pandas as pd
import numpy as np
from collections import deque
from library import *
```

# gym - copy-v0

```python
env1 = environments.environment_gym('Copy-v0')
env = environments.env_wrapper(env1)
n_actions = 20
```

```python
#random policy
policy = random()
Q = defaultdict(lambda: np.zeros(n_actions))
mean_scores, score, moving_window = play(env, Q, 10000, policy)
moving_window = moving_window[5:]
print('max',max(moving_window))
plt.plot(moving_window)
plt.show()
```

```python
policy = random()
Q_class = update_vanilla_Q(n_actions, gamma = 1.0)
mr = MC(env, policy, Q_class,500000, generate_episode, stopping = 25)
```

```python
policy = maximum()
mean_scores, score, moving_window = play(env, Q_class.Q, 10000, policy)
moving_window = moving_window[5:]
print('max',max(moving_window))
plt.plot(moving_window)
plt.show()
```

```python
eps = epsilon(eps_start = 1.0, eps_decay = 0.99, eps_min = 0.1)
policy = epsilon_greedy(eps,200)
Q_class = update_control_Q(n_actions,alpha = 0.2, gamma = 1.0)
mr = MC(env, policy, Q_class, 500000, generate_episode, stopping = 25)
```

```python
policy = maximum()
mean_scores, score, moving_window = play(env, Q_class.Q, 10000, policy)
moving_window = moving_window[5:]
print('max',max(moving_window))
plt.plot(moving_window)
plt.show()
```
