```python
import environments
import numpy as np
from collections import deque
```

```python
env1 = environments.environment_gym('Acrobot-v1')
env = environments.env_wrapper(env1,True)
```

```python
d = deque(maxlen = 1000)
```

```python
next_state = env.reset()
for i in range(0,1000):
    next_state, reward, is_done, _ = env.step(np.random.randint(env.action_size()))
    d.append([next_state, reward, is_done])
    if is_done:
        next_state = env.reset()
```
